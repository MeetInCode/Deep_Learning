```python
# prompt: import image and imagegenerator from keras.prepocessing

from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('dog.jpeg', target_size=(200,200))
```

```python
img
```

```python


# Create an ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',

)

```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
# prompt: write datagen,flow code

# Use the ImageDataGenerator to generate augmented images
i = 0
for batch in datagen.flow(input_batch, batch_size=1):
  # Save the augmented image
  image.save_img(f'augmented_image_{i}.jpg', batch[0])
  i += 1
  # Generate 5 augmented images
  if i > 5:
    break
```
